```python
df.drop(['salary_min', 'salary_max'], axis=1, inplace=True)
```

```python
df['date_posted'] = pd.to_datetime(df['date_posted'], errors='coerce')
df['date_posted'] = df['date_posted'].dt.tz_localize(None)
df['date_posted'] = df['date_posted'].fillna(pd.to_datetime('1900-01-01'))
df['date_posted'] = df['date_posted'].astype('datetime64[ns]')
```

```python
# df['date_posted'] = date_posted
df['date_posted_month'] = df['date_posted'].dt.month
df['date_posted_year'] = df['date_posted'].dt.year

df.tail(20)
```

```python
import re

def extract_seniority(title: str) -> str:
    if not title:
        return "unspecified"
    
    title_lower = title.lower()
    
    if any(x in title_lower for x in ["intern", "internship", "trainee"]):
        return "intern"
    elif any(x in title_lower for x in ["junior", "jr.", "jr ", "entry"]):
        return "junior"
    elif any(x in title_lower for x in ["mid", "mid-level", "intermediate"]):
        return "mid"
    elif any(x in title_lower for x in ["senior", "sr.", "sr ", "lead", "principal"]):
        return "senior"
    elif any(x in title_lower for x in ["manager", "head", "director", "vp"]):
        return "management"
    else:
        return "unspecified"
    

def extract_years_of_experience(desc: str) -> int:
    if not desc:
        return None
    
    desc_lower = desc.lower()
    
    patterns = [
        r'(\d+)\+?\s*(?:to\s*\d+)?\s*years?\s*(?:of\s*)?experience',
        r'minimum\s*(\d+)\s*years?',
        r'at\s*least\s*(\d+)\s*years?',
        r'(\d+)\s*\+\s*years?',
    ]
    
    for pattern in patterns:
        match = re.search(pattern, desc_lower)
        if match:
            return float(match.group(1))
    
    return None

```

```python
df['seniority'] = df["title"].apply(extract_seniority)

print(df['seniority'].value_counts())
```

```python
df['years_of_exp_req'] = df['description'].apply(extract_years_of_experience)

print(df['years_of_exp_req'].value_counts())
```
